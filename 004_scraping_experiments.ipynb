```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

base_url = "https://www.raddiscount.de/"
url = "https://www.raddiscount.de/Rennraeder.html"
r = requests.get(url,headers=headers) # fetch the website
```

```python
source = r.text # store its response in variables source

soup = BeautifulSoup(source, 'html.parser') # parse the webpage source into a bs4 soup object
print(soup.prettify())
```

```python
bikes = soup.find_all("td", {"class":"phead"})
#print(bikes)

for bike in bikes:
    print(bike)
    if "Bianchi" in bike.get_text():
        print(bike)
        page = bike.find("a").get("href")
        print("----")
        print(f"{base_url}{page}")
        break
```
